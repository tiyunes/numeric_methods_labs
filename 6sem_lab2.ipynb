```python
def getAnalytPulse(h, C, N):
  tau = C * h / a

  x_ranges = int(l / h)
  t_ranges = int(N / tau) - 1
  x = np.array([0.0] * x_ranges)
  t = [0] * t_ranges
  t_range = np.array([0.0] * t_ranges)

  for i in range(x_ranges):
    x[i] = S(i * h).round(5)
    
  for i in range(1, t_ranges):  
    t_range[i] = S(i * tau).round(5)
    t[i] = i

  u_analyt = np.array([[0.0] * x_ranges] * t_ranges)
  u_analyt[0][0] = 1

  for i in range(t_ranges):
    if (i != 0):
      u_analyt[i][0] = 0

  for i in range(x_ranges):
    if(x[i] < 1):
      u_analyt[0][i] = 1
    elif (x[i] >= 1):
      u_analyt[0][i] = 0

  for i in range(1, t_ranges):
    for j in range(1, x_ranges):
      if (x[j] < a * t_range[i]):
        u_analyt[i][j] = mu(t_range[i] - x[j] / a)
      else:
        u_analyt[i][j] = phi(x[j] - a * t_range[i])
  
  return u_analyt
```

```python
def getAnalytBell(h, C, N):
  tau = C * h / a

  x_ranges = int(l / h)
  t_ranges = int(N / tau) - 1
  x = np.array([0.0] * x_ranges)
  t = [0] * t_ranges
  t_range = np.array([0.0] * t_ranges)

  for i in range(x_ranges):
    x[i] = S(i * h).round(5)
    
  for i in range(1, t_ranges):  
    t_range[i] = S(i * tau).round(5)
    t[i] = i

  u_analyt_bell = np.array([[0.0] * x_ranges] * t_ranges)
  u_analyt_bell[0][0] = 0

  for i in range(t_ranges):
    if (i != 0):
      u_analyt_bell[i][0] = 0

  for i in range(x_ranges):
    u_analyt_bell[0][i] = phi_bell(x[i])

  for i in range(1, t_ranges):
    for j in range(1, x_ranges):
      if (x[j] < a * t_range[i]):
        # u_analyt[i][j] = mu(t_range[i] - x[j] / a)
        u_analyt_bell[i][j] = mu_bell(t_range[i] - x[j] / a)
      else:
        # u_analyt[i][j] = phi(x[j] - a * t_range[i])
        u_analyt_bell[i][j] = phi_bell(x[j] - a * t_range[i])

  return u_analyt_bell
```

```python
def getAnalytParabola(h, C, N):
  tau = C * h / a

  x_ranges = int(l / h)
  t_ranges = int(N / tau) - 1
  x = np.array([0.0] * x_ranges)
  t = [0] * t_ranges
  t_range = np.array([0.0] * t_ranges)

  for i in range(x_ranges):
    x[i] = S(i * h).round(5)
    
  for i in range(1, t_ranges):  
    t_range[i] = S(i * tau).round(5)
    t[i] = i

  u_analyt_parabola = np.array([[0.0] * x_ranges] * t_ranges)
  u_analyt_parabola[0][0] = 0

  for i in range(t_ranges):
    if (i != 0):
      u_analyt_parabola[i][0] = 0

  for i in range(x_ranges):
    u_analyt_parabola[0][i] = phi_parabola(x[i])

  for i in range(1, t_ranges):
    for j in range(1, x_ranges):
      if (x[j] < a * t_range[i]):
        # u_analyt[i][j] = mu(t_range[i] - x[j] / a)
        u_analyt_parabola[i][j] = mu_parabola(t_range[i] - x[j] / a)
      else:
        # u_analyt[i][j] = phi(x[j] - a * t_range[i])
        u_analyt_parabola[i][j] = phi_parabola(x[j] - a * t_range[i])

  return u_analyt_parabola
```

```python
def getAnalytCos(h, C, N):
  tau = C * h / a

  x_ranges = int(l / h)
  t_ranges = int(N / tau) - 1
  x = np.array([0.0] * x_ranges)
  t = [0] * t_ranges
  t_range = np.array([0.0] * t_ranges)

  for i in range(x_ranges):
    x[i] = S(i * h).round(5)
    
  for i in range(1, t_ranges):  
    t_range[i] = S(i * tau).round(5)
    t[i] = i

  u_analyt_cos = np.array([[0.0] * x_ranges] * t_ranges)
  u_analyt_cos[0][0] = 0

  for i in range(t_ranges):
    if (i != 0):
      u_analyt_cos[i][0] = 0

  for i in range(x_ranges):
    u_analyt_cos[0][i] = phi_cos(x[i])

  for i in range(1, t_ranges):
    for j in range(1, x_ranges):
      if (x[j] < a * t_range[i]):
        # u_analyt[i][j] = mu(t_range[i] - x[j] / a)
        u_analyt_cos[i][j] = mu_cos(t_range[i] - x[j] / a)
      else:
        # u_analyt[i][j] = phi(x[j] - a * t_range[i])
        u_analyt_cos[i][j] = phi_cos(x[j] - a * t_range[i])

  return u_analyt_cos
```

```python
def getNumericLax(phi, h, C, N):
  tau = C * h / a
  
  x_ranges = int(l / h)
  t_ranges = int(N / tau) - 1
  x = np.array([0.0] * x_ranges)
  t = [0] * t_ranges
  t_range = np.array([0.0] * t_ranges)

  ipos = np.zeros(x_ranges)
  ineg = np.zeros(x_ranges)

  for i in range(0, x_ranges):
    ipos[i] = int(i + 1)
    ineg[i] = int(i - 1)

  ipos[x_ranges - 1] = 0
  ineg[0] = x_ranges - 1

  for i in range(x_ranges):
    x[i] = S(i * h).round(5)
    
  for i in range(1, t_ranges):  
    t_range[i] = S(i * tau).round(5)
    t[i] = i

  u = np.array([[0.0] * x_ranges] * t_ranges)
  # u_cross = np.array([[0.0] * x_ranges] * t_ranges)

  u[0][0] = 0
  # u_cross[0][0] = 1

  for i in range(t_ranges):
    u[i][0] = 0 # гранич условия
    # u_cross[i][0] = 0

  for i in range(x_ranges):
    u[0][i] = phi(x[i])

  for i in range(1, t_ranges):
    for j in range(1, x_ranges):
      if (j < (x_ranges - 1)):
        u[i][j] = u[i - 1][j] + C * (C / 2) * (u[i - 1][j + 1] - 2 * u[i - 1][j] + u[i - 1][j - 1]) - 0.5 * C * (u[i - 1][j + 1] - u[i - 1][j - 1])
      elif (j == (x_ranges - 1)):
        u[i][j] = u[i - 1][j] + C * (C / 2) * (u[i - 1][0] - 2 * u[i - 1][j] + u[i - 1][j - 1]) - 0.5 * C * (u[i - 1][0] - u[i - 1][j - 1])
      
      # if (j < (x_ranges - 1)):
        # u_cross[i][j] = # u_cross[i - 1][j] - 0.5 * C * (# u_cross[i - 1][j + 1] - # u_cross[i - 1][j - 1])
  
  return u
```

```python
def getError(u, u_analyt, h, C, N):
  tau = C * h / a
  x_ranges = int(l / h)
  t_ranges = int(N / tau) - 1

  h_angle = np.array([[0.0] * x_ranges] * t_ranges)
  max_error = np.array([0.0] * t_ranges)
  for i in range(t_ranges):
    max_error[i] = 0
    for j in range(x_ranges):
      h_angle[i][j] = abs(u_analyt[i][j] - u[i][j])
      max_error[i] = max(max_error[i], h_angle[i][j])

  return max_error
```

```python
def plot(u1, u2, u_analyt, h1, h2, error1, error2, C, N, scheme_name):
  tau1 = C * h1 / a
  tau2 = C * h2 / a

  eps1 = error1[-1]
  eps2 = error2[-1]
  
  x_ranges1 = int(l / h1)
  t_ranges1 = int(N / tau1) - 1
  x1 = np.array([0.0] * x_ranges1)
  t1 = [0] * t_ranges1
  t_range1 = np.array([0.0] * t_ranges1)

  for i in range(x_ranges1):
    x1[i] = S(i * h1).round(5)
    
  for i in range(1, t_ranges1):  
    t_range1[i] = S(i * tau1).round(5)
    t1[i] = i

  x_ranges2 = int(l / h2)
  t_ranges2 = int(N / tau2) - 1
  x2 = np.array([0.0] * x_ranges2)
  t2 = [0] * t_ranges2
  t_range2 = np.array([0.0] * t_ranges2)

  t_ranges = min(t_ranges1, t_ranges2) - 1
  t = [0] * t_ranges
  t_range = np.array([0.0] * t_ranges)
  if (t_ranges1 < t_ranges2):
    tau_ = tau1
  else:
    tau_ = tau2

  for i in range(x_ranges2):
    x2[i] = S(i * h2).round(5)
    
  for i in range(1, t_ranges2):  
    t_range2[i] = S(i * tau2).round(5)
    t2[i] = i

  for i in range(1, t_ranges):  
    t_range[i] = S(i * tau_).round(5)
    t[i] = i

  fig, ax = plt.subplots(2, figsize = (11, 7))

  line1, = ax[0].plot(x1, u1[0])
  line1.set_color("green")
  line1.set_label(f"numeric solution (angle) h1 = {h1}")
  # print("ok1")

  line2, = ax[0].plot(x2, u2[0])
  line2.set_color("black")
  line2.set_label(f"numeric solution (angle) h2 = {h2}")
  # print("ok2")

  line_analyt, = ax[0].plot(x2, u_analyt[0])
  line_analyt.set_color("blue")
  line_analyt.set_label("analytic solution")
  # print("ok3")

  line_error1, = ax[1].plot(t_range, error1[:t_ranges])
  eps1 = error1[0]
  line_error1.set_label(f"errors (h1 = {h1})")
  line_error1.set_color("green")
  # print("ok4")

  line_error2, = ax[1].plot(t_range, error2[:t_ranges])
  eps2 = error2[0]
  line_error2.set_label(f"errors (h2 = {h2})")
  line_error2.set_color("black")
  # print("ok5")

  # line_cross, = ax.plot(x, u_cross[0])
  # line_cross.set_color("red")
  # line_cross.set_label("numeric solution (half-cross)")

  ax[0].set_xlabel("x")
  ax[0].set_ylabel("u")
  ax[0].set_title(f"{scheme_name}, C = {C}")
  ax[0].set_ylim(-0.2, 1.3)
  if (eps2 != 0):
    ax[1].set_title(f" eps1 / eps2  = {(eps1 / eps2).round(3)}")
  annotation = ax[0].annotate(f"t = {t_range[0].round(3)}", xy = (5, 1), ha = "center")
  annotation.set_animated(True)

  hlegend0 = ax[0].legend()
  hlegend1 = ax[1].legend()

  def init():
    return line1, line2, line_analyt, line_error1, line_error2, annotation

  def update_u(frame, line1, line2, line_analyt, line_errror1, line_error2, x):
    line1.set_ydata(u1[frame])
    line2.set_ydata(u2[int(h1 / h2) * frame])
    line_error1.set_xdata(t_range[:frame])
    line_error1.set_ydata(error1[:frame])
    line_error2.set_xdata(t_range[:frame])
    line_error2.set_ydata(error2[:frame])
    line_analyt.set_ydata(u_analyt[int(h1 / h2) * frame])
    if (error2[frame] != 0):
      ax[1].set_title(f" eps1 / eps2  = {(error1[frame] / error2[frame]).round(3)}")


    annotation.set_text(f"t = {t_range[frame].round(3)}")
    # ax.annotate(f"t = {t_range[frame].round(3)}", xy = (5, 1), ha = "center")
    # line_cross.set_ydata(u_cross[frame])
    htext0 = hlegend0.get_texts()[0]
    htext1 = hlegend1.get_texts()[0]
    return line1, line2, line_analyt, line_error1, line_error2, htext0, htext1, annotation

  # t.pop()
  anim = animation.FuncAnimation(fig, update_u, frames = t, init_func = init, fargs = (line1, line2, line_analyt, line_error1, line_error2, x1), interval = 60, blit = True, repeat = False)
  # plt.ioff()
  IPython.display.display(anim)
```

```python
def plotNumericLax(phi, mu, h1, h2, C, N):
  tau1 = C * h1 / a
  tau2 = C * h2 / a

  x_ranges1 = int(l / h1)
  t_ranges1 = int(N / tau1)
  x1 = np.array([0.0] * x_ranges1)
  t1 = [0] * t_ranges1
  t_range1 = np.array([0.0] * (t_ranges1))

  for i in range(x_ranges1):
    x1[i] = S(i * h1).round(5)
    
  for i in range(t_ranges1):  
    t_range1[i] = S(i * tau1).round(5)
    t1[i] = i

  x_ranges2 = int(l / h2)
  t_ranges2 = int(N / tau2) - 1
  x2 = np.array([0.0] * x_ranges2)
  t2 = [0] * t_ranges2
  t_range2 = np.array([0.0] * t_ranges2)

  t_ranges = min(t_ranges1, t_ranges2)
  t = [0] * t_ranges
  t_range = np.array([0.0] * t_ranges)

  for i in range(x_ranges2):
    x2[i] = S(i * h2).round(5)

  for i in range(1, t_ranges2):  
    t_range2[i] = S(i * tau2).round(5)
    t2[i] = i
  t2.pop()
  if (t_ranges1 < t_ranges2):
    tau_ = tau1
  else:
    tau_ = tau2

  for i in range(1, t_ranges):  
    t_range[i] = S(i * tau_).round(5)
    t[i] = i

  u1 = np.array([[0.0] * x_ranges1] * 2)
  u1[0][0] = 0

  for i in range(x_ranges1):
    u1[0][i] = phi(x1[i])

  u2 = np.array([[0.0] * x_ranges2] * 2)
  u2[0][0] = 0

  for i in range(x_ranges2):
    u2[0][i] = phi(x2[i])

  u_analyt = np.array([[0.0] * x_ranges2] * 2)
  u_analyt[0][0] = 0

  u_analyt1 = np.array([[0.0] * x_ranges1] * 2)
  u_analyt1[0][0] = 0

  for i in range(x_ranges2):
    u_analyt[0][i] = phi(x2[i])

  for i in range(x_ranges1):
    u_analyt1[0][i] = phi(x1[i])

  h_angle1 = np.array([0.0] * x_ranges1)
  h_angle2 = np.array([0.0] * x_ranges2)
  # h_angle2 = np.array([0.0] * x_ranges1)

  max_error1 = np.array([0.0] * (t_ranges1 + 5))
  max_error2 = np.array([0.0] * (t_ranges2 + 5))
  # max_error2 = np.array([0.0] * t_ranges1)

  print("t_ranges2: ", t_ranges2)
  fig, ax = plt.subplots(2, figsize = (11, 7))

  line1, = ax[0].plot(x1, u1[0])
  line1.set_color("green")
  line1.set_label(f"numeric solution (angle) h1 = {h1}")

  line2, = ax[0].plot(x2, u2[0])
  line2.set_color("black")
  line2.set_label(f"numeric solution (angle) h2 = {h2}")

  line_analyt, = ax[0].plot(x2, u_analyt[0])
  line_analyt.set_color("blue")
  line_analyt.set_label("analytic solution")

  line_error1, = ax[1].plot(t_range, max_error1[:t_ranges])
  max_error1[0] = abs(max(u1[0] - u_analyt1[0]))
  eps1 = max_error1[0]
  line_error1.set_label(f"errors (h1 = {h1})")
  line_error1.set_color("green")

  line_error2, = ax[1].plot(t_range, max_error2[:t_ranges])
  eps2 = abs(max(u2[0] - u_analyt[0]))
  max_error2[0] = abs(max(u2[0] - u_analyt[0]))
  line_error2.set_label(f"errors (h2 = {h2})")
  line_error2.set_color("black")

  ax[0].set_xlabel("x")
  ax[0].set_ylabel("u")
  ax[0].set_title(f"Upwind, C = {C}")
  ax[1].set_ylim(0, 1)
  if (eps2 != 0):
    ax[1].set_title(f" eps1 / eps2  = {(eps1 / eps2).round(3)}")
  annotation = ax[0].annotate(f"t = {t_range[0].round(3)}", xy = (5, 1), ha = "center")
  annotation.set_animated(True)

  hlegend0 = ax[0].legend()
  hlegend1 = ax[1].legend()

  def init():
    return line1, line2, line_analyt, line_error1, line_error2, annotation

  def update_u(frame, line1, line2, line_analyt, line_errror1, line_error2, x):

    # print("frame ok2: ", frame)
    u2[(1 - frame % 2) % 2][0] = 0
    u_analyt[(1 - frame % 2) % 2][0] = 0
    for j in range(1, x_ranges2):
      if (j < (x_ranges2 - 1)):
        u2[(1 - frame % 2) % 2][j] = u2[frame % 2][j] + C * (C / 2) * (u2[frame % 2][j + 1] - 2 * u2[frame % 2][j] + u2[frame % 2][j - 1]) - 0.5 * C * (u2[frame % 2][j + 1] - u2[frame % 2][j - 1])
      elif (j == (x_ranges2 - 1)):
        u2[(1 - frame % 2) % 2][j] = u2[frame % 2][j] + C * (C / 2) * (u2[frame % 2][0] - 2 * u2[frame % 2][j] + u2[frame % 2][j - 1]) - 0.5 * C * (u2[frame % 2][0] - u2[frame % 2][j - 1])

      if (x2[j] < a * t_range2[frame]):
        u_analyt[(1 - frame % 2) % 2][j] = mu(t_range2[frame] - x2[j] / a)
      else:
        u_analyt[(1 - frame % 2) % 2][j] = phi(x2[j] - a * t_range2[frame])


    if (frame % (int(h1 / h2)) == 0 and (frame < (t_ranges2 - 1))):
      u1[(1 - (frame // int(h1/h2)) % 2) % 2][0] = 0
      u_analyt1[(1 - (frame // int(h1/h2)) % 2) % 2][0] = 0
      for j in range(1, x_ranges1):
        if (j < (x_ranges1 - 1)):
          u1[(1 - (frame // int(h1/h2)) % 2) % 2][j] = u1[(frame // int(h1/h2)) % 2][j] + C * (C / 2) * (u1[(frame // int(h1/h2)) % 2][j + 1] - 2 * u1[(frame // int(h1/h2)) % 2][j] + u1[(frame // int(h1/h2)) % 2][j - 1]) - 0.5 * C * (u1[(frame // int(h1/h2)) % 2][j + 1] - u1[(frame // int(h1/h2)) % 2][j - 1])
        elif (j == (x_ranges1 - 1)):
          u1[(1 - (frame // int(h1/h2)) % 2) % 2][j] = u1[(frame // int(h1/h2)) % 2][j] + C * (C / 2) * (u1[(frame // int(h1/h2)) % 2][0] - 2 * u1[(frame // int(h1/h2)) % 2][j] + u1[(frame // int(h1/h2)) % 2][j - 1]) - 0.5 * C * (u1[(frame // int(h1/h2)) % 2][0] - u1[(frame // int(h1/h2)) % 2][j - 1])

        if (x1[j] < a * t_range2[frame]):
          u_analyt1[(1 - (frame // int(h1/h2)) % 2) % 2][j] = mu(t_range2[frame] - x1[j] / a)
        else:
          u_analyt1[(1 - (frame // int(h1/h2)) % 2) % 2][j] = phi(x1[j] - a * t_range2[frame])
    
      # if (frame % (int(h1 / h2)) == 0):
        max_error1[frame // (int(h1/h2))] = 0
        for j in range(1, x_ranges1):
          h_angle1[j] = abs(u_analyt1[(1 - (frame // int(h1/h2)) % 2) % 2][j] - u1[(1 - (frame // int(h1/h2)) % 2) % 2][j])
          max_error1[frame // (int(h1/h2))] = max(max_error1[frame // (int(h1/h2))], h_angle1[j])

    # print("frame ok3: ", frame)
    max_error2[frame] = 0
    for j in range(1, x_ranges2):
      h_angle2[j] = abs(u_analyt[(1 - frame % 2) % 2][j] - u2[(1 - frame % 2) % 2][j])
      max_error2[frame] = max(max_error2[frame], h_angle2[j])
    # for j in range(1, x_ranges1):
    #   h_angle2[j] = abs(u_analyt1[frame % 2][j] - u2[((frame) * int(h1 / h2)) % 2][j])
    #   max_error2[frame] = max(max_error2[frame], h_angle2[j])

    # print("u1 len ", len(u1[(1 - frame % 2) % 2]))
    line1.set_ydata(u1[(1 - (frame // int(h1/h2)) % 2) % 2])
    # line2.set_ydata(u2[((frame + 1) * int(h1 / h2)) % 2])
    # print("u2 len ", len(u2[(1 - frame % 2) % 2]))
    line2.set_ydata(u2[(1 - frame % 2) % 2])
    # print("t_range1: ", t_range1[:frame])
    # print("t_range2 // int len ", len(t_range2[:(frame // int(h1/h2))]))
    # print("max error 1 // int len ", len(max_error1[:(frame // int(h1/h2))]))
    # print("frame ok4: ", frame)
    if (frame < (t_ranges2 - 1)):
      line_error1.set_xdata(t_range1[:(frame // int(h1/h2))])
      line_error1.set_ydata(max_error1[:(frame // int(h1/h2))])
    # print("error 1 ", (max_error1[:(frame // int(h1/h2))]))
    # print("frame ", frame)


    # line_error2.set_xdata(t_range2[:frame])
    # line_error2.set_ydata(max_error2[:frame])

    line_error2.set_xdata(t_range2[:frame:int(h1/h2)])
    line_error2.set_ydata(max_error2[:frame:int(h1/h2)])

    # line_analyt.set_ydata(u_analyt[((frame + 1) * int(h1 / h2)) % 2])
    line_analyt.set_ydata(u_analyt[(1 - frame % 2) % 2])
    if (max_error2[frame] != 0 and frame % int(h1 / h2) == 0):
      ax[1].set_title(f" eps1 / eps2  = {(max_error1[frame // int(h1/h2)] / max_error2[frame]).round(3)}")
    # print("frame ok5: ", frame)

    annotation.set_text(f"t = {t_range2[frame].round(3)}")
    # ax.annotate(f"t = {t_range[frame].round(3)}", xy = (5, 1), ha = "center")
    # line_cross.set_ydata(u_cross[frame])
    htext0 = hlegend0.get_texts()[0]
    htext1 = hlegend1.get_texts()[0]
    return line1, line2, line_analyt, line_error1, line_error2, htext0, htext1, annotation

  # t.pop()
  anim = animation.FuncAnimation(fig, update_u, frames = t2, init_func = init, fargs = (line1, line2, line_analyt, line_error1, line_error2, x1), interval = 80, blit = True, repeat = False)
  # plt.ioff()
  IPython.display.display(anim)
```
